# Comparative Graphs of various Sorting Algorithms

**For Source Codes and Results, refer to the Graphs folder.**

```python
#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reading Input-Size v/s Time data of Bubble Sort
time_bubbleSort = pd.read_csv('Graphs/Results(BSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_bubbleSort)
plt.title('Bubble Sort')
```

```python
#Reading Input-Size v/s Time data of Selection Sort
time_selectionSort = pd.read_csv('Graphs/Results(SSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_selectionSort)
plt.title('Selection Sort')
```

```python
#Reading Input-Size v/s Time data of Insertion Sort
time_insertionSort = pd.read_csv('Graphs/Results(ISort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_insertionSort)
plt.title('Insertion Sort')
```

```python
#Reading Input-Size v/s Time data of Merge Sort
time_mergeSort = pd.read_csv('Graphs/Results(MSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_mergeSort)
plt.title('Merge Sort')
```

```python
#Reading Input-Size v/s Time data of Quick Sort
time_quickSort = pd.read_csv('Graphs/Results(QSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_quickSort)
plt.title('Quick Sort')
```

```python
#Reading Input-Size v/s Time data of Heap Sort
time_heapSort = pd.read_csv('Graphs/Results(HSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_heapSort)
plt.title('Heap Sort')
```

```python
#Reading Input-Size v/s Time data of Count Sort
time_countsort = pd.read_csv('Graphs/Results(CSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_countsort)
plt.title('Count Sort')
```

```python
#Reading Input-Size v/s Time data of Radix Sort
time_radixsort = pd.read_csv('Graphs/Results(RSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_radixsort)
plt.title('Radix Sort')
```

```python
#Reading Input-Size v/s Time data of Radix Sort
time_bucketsort = pd.read_csv('Graphs/Results(BucketSort).csv')
```

```python
sns.lineplot(x='N',y='Time',data=time_bucketsort)
plt.title('Bucket Sort')
```

**All combined**

```python
plt.figure(figsize=(12,8))
sns.lineplot(time_bubbleSort['N'],time_bubbleSort['Time'])
sns.lineplot(time_selectionSort['N'],time_selectionSort['Time'])
sns.lineplot(time_insertionSort['N'],time_insertionSort['Time'])
sns.lineplot(time_mergeSort['N'],time_mergeSort['Time'])
sns.lineplot(time_quickSort['N'],time_quickSort['Time'])
sns.lineplot(time_heapSort['N'],time_heapSort['Time'])
sns.lineplot(time_countsort['N'],time_countsort['Time'])
sns.lineplot(time_radixsort['N'],time_radixsort['Time'])
sns.lineplot(time_bucketsort['N'],time_bucketsort['Time'])
plt.legend(['Bubble Sort','Selection Sort','Insertion Sort','Merge Sort','Quick Sort','Heap Sort','Count Sort','Radix Sort','Bucket Sort'])
plt.title('Comparitive Graph of All Algorithms Together')
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(time_bubbleSort['N'],time_bubbleSort['Time'])
sns.lineplot(time_selectionSort['N'],time_selectionSort['Time'])
sns.lineplot(time_insertionSort['N'],time_insertionSort['Time'])
plt.legend(['Bubble Sort','Selection Sort','Insertion Sort'])
plt.title('Comparitive Graph for Algorithms with Quadratic Running Time')
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(time_mergeSort['N'],time_mergeSort['Time'])
sns.lineplot(time_quickSort['N'],time_quickSort['Time'])
sns.lineplot(time_heapSort['N'],time_heapSort['Time'])
plt.legend(['Merge Sort','Quick Sort','Heap Sort'])
plt.title('Comparitive Graph for Algorithms with NlogN Running Time')
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(time_countsort['N'],time_countsort['Time'])
sns.lineplot(time_radixsort['N'],time_radixsort['Time'])
sns.lineplot(time_bucketsort['N'],time_bucketsort['Time'])
plt.legend(['Count Sort','Radix Sort','Bucket Sort'])
plt.title('Comparitive Graph for Algorithms with Linear Running Time')
```
